```python
import numpy as np
import pandas as pd
import sklearn as sk
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("austin_weather.csv")
```

```python
data = data.drop(["Events", "Date", "SeaLevelPressureLowInches"], axis = 1)
```

```python
data["PrecipitationSumInches"] = data["PrecipitationSumInches"].replace("T", 0).astype(float)
data = data.replace("-", 0).astype(float)
```

```python
X = data.drop(["PrecipitationSumInches"], axis = 1)
Y = data["PrecipitationSumInches"].values.reshape(-1,1)
```

```python
clf = LinearRegression()
clf.fit(X, Y)
```

```python
day_index = 1000
days = np.arange(len(Y))
```

```python
print("The Precipitation trend Graph")
plt.scatter(days, Y.flatten())
plt.scatter(days[day_index], Y.flatten()[day_index], c="red")
plt.xlabel("Days")
plt.ylabel("Precipitation in Inches")
plt.show()
x_vis = X.filter(['TempAvgF', 'DewPointAvgF', 'HumidityAvgPercent', 'SeaLevelPressureAvgInches', 'VisibilityAvgMiles', 'WindAvgMPH'])
```

```python
print("The Graphs of precipitation vs attributes trends graph")
for i in range(x_vis.columns.size):
    plt.subplot(3, 2, i+1)
    plt.scatter(days, x_vis[x_vis.columns.values[i]], color='g')
    plt.scatter(days[day_index], x_vis[x_vis.columns.values[i]][day_index], color='r')
    plt.title(x_vis.columns.values[i])
plt.show()
```
